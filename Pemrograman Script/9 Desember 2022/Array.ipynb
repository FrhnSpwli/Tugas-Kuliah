```python
#Nomor 1
#Deklarasi numpy array kosong berukuran 2 x 3, beri array tersebut nama array1
import numpy as np
array1 = np.empty([2,3], dtype = int)
print(array1)
```

```python
#Nomor 2
#Deklarasi numpy array berisi angka 0, berukuran 100 x 150, beri array tersebut nama array2
array2 = np.zeros([100,150])
print(array2)
```

```python
#Nomor 3
#Deklarasi numpy array berisi angka 1, berukuran 150 x 100, beri array tersebut nama array3
array3 = np.ones([150,100])
print(array3)  
```

```python
#Nomor 4
#Cari type data dari array1, array2, dan array3
print(array1.dtype)
print(array2.dtype)
print(array3.dtype)
```

```python
#Nomor 5
#Cari ukuran (shape) dari array 2
print(array2.shape)
```

```python
#Nomor 6
#Putar array3 sehingga menjadi berukuran 100 x 150
array3T = np.transpose(array3)
print(array3T)
print(array3T.shape)
```

```python
#Nomor 7
#Satukan array2 dan array3 menjadi sebuah array berukuran 100 x 300, beri array tersebut nama array4
array4 = np.concatenate((array2, array3T), axis=1)
print(array4)
print(array4.shape)
```

```python
#Nomor 8
#Deklarasikan sebuah array kososng berukuran 5 x 7, beri nama array5. Array dipersiapkan untuk menyimpan nilai-nilai bertype float
array5 = np.empty([5,7], dtype=float)
print(array5)
```

```python
#Nomor 9
#Gunakan loop (for atau while) untuk mem-printout (display) setiap baris dari array2
print("Menggunakan loop for")
for i in array2:
    print(i)

print("Menggunakan loop while")
i = 0
while i < len(array2):
    print(array2[i])
    i += 1
```

```python
#Nomor 10
#Gunakan array1 untuk menunjukkan perbedaan bila melakukan shallow copy dan deep copy

#Shallow Copy
shallow = array1.view()
print(shallow)

#Deep Copy
deep = array1.copy()
print(deep)

array1[0][0] = 1   
print(array1[0][0])
print(shallow[0][0])
print(deep[0][0])
```
