```python
import pandas as pd

df = pd.read_csv('dinkes.csv')
print(df.to_string())
```

```python
#Nomor 1
#print kategori puskesmas
print(df['kategori_puskesmas'].value_counts())
```

```python
#Nomor 2
#Jumlah puskesmas di Kabupaten Cianjur
cianjur = df.where(
    (df['nama_kabupaten_kota'] == 'KABUPATEN CIANJUR') &
    (df['tahun'] == 2021)
).dropna()
cianjur
```

```python
jumlah_puskesmas = cianjur['jumlah_puskesmas'].sum()
print("Jumlah Puskesmas di Kab. Cianjur : ", jumlah_puskesmas)
```

```python
#Nomor 3
bogor2021 = df.where(
    (df['nama_kabupaten_kota'] == 'KABUPATEN BOGOR') &
    (df['tahun'] == 2021) & (df['kategori_puskesmas'] == 'PUSKESMAS RAWAT INAP')
).dropna()
jumlah2021 = bogor2021['jumlah_puskesmas'].sum()

bogor2019 = df.where(
    (df['nama_kabupaten_kota'] == 'KABUPATEN BOGOR') &
    (df['tahun'] == 2019) & (df['kategori_puskesmas'] == 'PUSKESMAS RAWAT INAP')
).dropna()
jumlah2019 = bogor2019['jumlah_puskesmas'].sum()

print("Jumlah Puskesmas di Kab. Bogor tahun 2019 : ", jumlah2019)
print("Jumlah Puskesmas di Kab. Bogor tahun 2021 : ", jumlah2021)
print("Pertambahan jumlah puskesmas di Kab. Bogor tahun 2021 dibandingkan tahun 2019 : ", jumlah2021-jumlah2019)
```
